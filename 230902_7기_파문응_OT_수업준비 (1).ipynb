# CRISP-DM 방법론

- 1단계 비즈니스 이해 (Business Understanding)
    - 현재 상황 평가
    - 데이터 마이닝 목표 결정
    - 프로젝트 계획 수립
----    
- 2단계 데이터에 대한 이해 (Data Understanding)
    - 데이터 설명
    - 데이터 탐색
    - 데이터 품질 확인
----
- 3단계 데이터 준비 (Data Preparation)
    - 데이터 선택
    - 데이터 정제
    - 필수데이터 구성
    - 데이터 통합
----
- 4단계 모델링 (Modeling)
    - 모델링 기법 선정
    - 테스트 디자인 생성
    - 모델 생성
    - 모델 평가
- 5단계 평가 (Evaluation)
    - 결과 평가
    - 프로세스 검토
    - 다음 단계 결정
----
- 6단계 배포 (Deployment)
    - 배포 계획
    - 모니터링 및 유지 관리 계획
    - 최종 보고서 작성
    - 프로젝트 검토

## SEMMA 방법론
---
- Sampling (데이터추출)
- Exploration (데이터 탐색)
- Modification (변수가공)
- Modeling (모델구축)
- Assessment (모델평가)
