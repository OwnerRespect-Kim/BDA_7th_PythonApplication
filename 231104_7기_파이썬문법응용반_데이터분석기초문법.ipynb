### 행이 많아지는 것과, 열이 많아지는 것 둘이 의미하는 것은 어떤 것일까?
- 행이 많아지는 것? 관측수
- 열이 많아지는 것? 변수

- nclass, math, english  변수가 총 3개
- 1, 50, 60 기타 등등
- 행의 개념
- x, y, z
- 1, 50, 60

- 2x + 4y + 3z = 10
- 4x + 8y + 6z = 20 (x2)
- 3x + 5y + 4z = 5
-----
- 열의 개념
- nclass, math, english, science , mean, str
- x, y, z, k, e, f

- 열이 많아지면 우리가 해결해야 하는 미지수들이 많아지는 것
- 차원이 복잡해지는 것
- 설명할 수 있는 차원들이 많아지면서 데이터가 복잡하게 만들어 진다.

### 데이터 정렬하기!
- 내림차순과 오름차순

```python
df_sp
```

```python
# 수학을 정렬의 기준으로 잡자.
print(df_sp.sort_values('math')) #기본 디폴트는 오름차순이다!

# 내림차순으로 정렬하고 싶다.
print(df_sp.sort_values('math', ascending=False))
```

```python
# 하나의 변수만 내림, 오름 차순이 아니라 두 가지 이상의 변수를 사용하여 정렬하는 경우
print(df_sp.sort_values(['math','english']))

print(df_sp.sort_values(['english','math']))

# 하나는 오름차순, 하나는 내림차순 두 가지 조건을 다르게 하는 경우


print(df_sp.sort_values(['english','math'], ascending=[True, False]))
```

### groupby
- 집단별로 묶어서 통계치 등을 함께 보는 것
- 반별로의 수학평균, 반별로의 영어 평균

- groupby 데이터를 묶는 개념 + 내가 원하는 값( 통계치가 필요하다 ) agg 함수랑 같이 많이 사용한다.

```python
df_sp.groupby('nclass').agg(mean_math =('math','mean')) # 변수와, 통계치 쓰면 된다.
```

```python
df_sp.groupby('nclass').agg(mean_math =('math','sum'))
```

```python
df_sp.groupby('nclass').agg(mean_math =('math','max'))
```

```python
df_sp.groupby('nclass').agg(mean_math =('math','mean'), # 수학 평균
                           sum_english =('english','sum'), # 영어 전체 성적 합산
                           nclass_count = ('nclass','count')) #count
```
